```python
from langchain_upstage import ChatUpstage
llm = ChatUpstage()
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt_template1 = PromptTemplate.from_template(
    """
   Step1: 다음 문장으로 시작하는 소설의 도입부를 한 문단 전개해줘. 이야기는 우리 둘이 같이 쓰는 거니까 너 혼자 끝내면 안 돼. 꼭 1개 문단으로만 써줘. {text}

   Step2: 다음 문단으로 이어지는 선택지를 가능한 다양한 형식으로 4개 작성해줄래? 이야기를 너 혼자 끝내지 말고 한 문단씩만 작성해줘. 이야기는 우리 둘이 같이 쓰는 거야. 꼭 1개 문단으로만 써줘.
        - 선택1
        - 선택2
        - 선택3
        - 선택4
   
"""
)
chain = prompt_template1 | llm | StrOutputParser()
text1 = "손가락이 두 개인 초록 개구리가 연못에 있어. 연못의 이웃으로는 흰 오리 한 마리가 있어. 개구리와 오리의 이름을 지어주고 이야기 시작해줘. 우선 한 문단 정도 적어줄래?"
select_1st = chain.invoke({"text": text1})
select_1st
```

```python
selected_1st = 3
if selected_1st == 4:
    next_input = select_1st[select_1st.find("Step1")+7:select_1st.find("Step2")] + select_1st[select_1st.find("선택" + str(selected_1st))+5:]
else:
    next_input = select_1st[select_1st.find("Step1")+7:select_1st.find("Step2")] + select_1st[select_1st.find("선택" + str(selected_1st))+5:select_1st.find("선택" + str(selected_1st+1))]
next_input
```
